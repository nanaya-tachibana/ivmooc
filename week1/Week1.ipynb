```python
from graph_tool.all import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nodes = pd.read_csv("florentine_node.csv", sep=' ')
edges = pd.read_csv("florentine_edge.csv", sep=' ')
```

```python
nodes
```

```python
edges
```

```python
g = Graph(directed=False)
vprop_label = g.new_vertex_property("string")
vprop_wealth = g.new_vertex_property("int")
vprop_priorates = g.new_vertex_property("vector<float>")
max_wealth = nodes["wealth"].max()
min_wealth = nodes["wealth"].min()
max_priorates = nodes["priorates"].max()
min_priorates = nodes["priorates"].min()
for node in nodes.itertuples():
    _, id, label, wealth, totalities, priorates = node
    id -= 1
    v = g.add_vertex()
    vprop_label[v] = label
    vprop_wealth[v] = ((wealth - min_wealth) / max_wealth + .1) * 100
    vprop_priorates[v] = plt.get_cmap("gist_heat")((priorates - min_priorates) / max_priorates / 2)
    
eprop_color = g.new_edge_property("string")
for edge in edges.itertuples():
    _, source, target, marriage, bussiness = edge
    source -= 1
    target -= 1
    if marriage == "T":
        e = g.add_edge(g.vertex(source), g.vertex(target))
        eprop_color[e] = "pink"
    if bussiness == "T":
        e = g.add_edge(g.vertex(source), g.vertex(target))
        eprop_color[e] = "gray"
```

```python
graph_draw(g, vertex_text=vprop_label, vertex_text_position=6, vertex_font_size=12,
           vertex_size=vprop_wealth, vertex_fill_color=vprop_priorates,
           vertex_color=[0, 0, 0, 0], edge_color=eprop_color,
           output_size=(800, 800), inline=True, fmt="pdf", output="florentine.pdf")
```
